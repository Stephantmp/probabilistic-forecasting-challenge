```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Define the pinball loss function
def pinball_loss(error, quantile):
    return np.where(error > 0, 2*(1 - quantile) * error, 2*quantile * -error)

# Define the quantiles
quantiles = [0.025, 0.25, 0.5, 0.75, 0.975]

# Generate a range of errors (from -2 to 2)
errors = np.linspace(-2, 2, 400)

# Set the Seaborn style
sns.set(style="whitegrid")

# Define the colors and line styles
new_colors = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
line_styles = ['-', '--', '-.', ':', (0, (3, 5, 1, 5))]

# Create the figure
plt.figure(figsize=(12, 7))

# Plot each quantile
for quantile, color, line_style in zip(quantiles, new_colors, line_styles):
    sns.lineplot(x=errors, y=pinball_loss(errors, quantile), label=f'$\\tau = {quantile}$', 
                 color=color, linestyle=line_style)

# Enhancing readability
plt.title('Pinball Loss for Different Quantiles ($\\tau$)', fontsize=16)
plt.xlabel('Error ($q^{\\tau} - y$)', fontsize=14)
plt.ylabel('Pinball Loss ($L_{\\tau}(q^{\\tau}, y)$)', fontsize=14)

# Position the legend at the top center, stacked
plt.legend(loc='upper center', fontsize=12, bbox_to_anchor=(0.5, 1), ncol=1)

# Show the plot
plt.show()

```
