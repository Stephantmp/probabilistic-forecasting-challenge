## Evaluation of final test set

```python
def evaluate_per_year(last_year=3, start_date='2023-09-01', end_date='2023-09-30'):
    daxdata=get_DAX.get(last_year)

    baseline_model = {
        'name': 'Baseline Model',
        'function': DAX_baseline.DAX_baseline
    }
    quantile_regression_model = {
        'name': 'Quantile Regression',
        'function': DAX_quantile_regression.DAX_quantile_regression
    }
    lstm_model={
        'name': 'LSTM',
        'function': DAX_LSTM.DAX_LSTM
    }
    xgboost_model = {
        'name': 'XGBoost',
        'function': DAX_XGBoost.run_model
    }
    
    evaluations, grouped_scores = evaluate_dax.evaluate(lstm_model, baseline_model,quantile_regression_model, xgboost_model, df=daxdata, start_date=start_date, end_date=end_date)
    return evaluations, grouped_scores
```

### Run evaluation for time horizon of 3 years

```python
evaluations_3,grouped_scores_3=evaluate_per_year(last_year=3, start_date='2023-11-15', end_date='2024-02-14')
```

```python
# group scores per mean for each year
grouped_scores_3.groupby('model').mean()
```

### Include ensemble model

```python
from functions.evaluation import evaluate_horizon
from functions.naive_ensemble import ensemble

df_models = [evaluations_3[0], evaluations_3[3]]  # Your DataFrames containing of LSTM and XGBoost
weights = [0.56, 0.44]  # Weights for each model's forecast
ensemble_df = ensemble(df_models, weights)
ensemble_df = ensemble_df.reset_index(drop=True)
```

```python
for index, row in ensemble_df.iterrows():
        horizon_days = int(row['horizon'].split()[0])
        ret_str = f'lag_ret{horizon_days}'
        observation = row[ret_str]
        quantile_preds = row[['q0.025', 'q0.25', 'q0.5', 'q0.75', 'q0.975']]
        score = evaluate_horizon(quantile_preds.values, observation)
        ensemble_df.at[index, 'score'] = score
```

### Evaluate per horizon
Also calculate the actual mean of the lag_retX for each horizon

```python
for evaluation_i in evaluations_3:
    # Print the model name from the first row
    model_name = evaluation_i['model'].iloc[0]
    print(model_name)

    # Calculate the mean for the specified columns, grouped by 'horizon'
    group_means = evaluation_i[['horizon', 'q0.025', 'q0.25', 'q0.5', 'q0.75', 'q0.975', 'score']].groupby('horizon').mean()

    # Assuming 'horizon' is a column where each value indicates the day, e.g., '1 day', '2 days', etc.
    # And you want to add the corresponding 'lag_retX' value to each group
    for horizon in evaluation_i['horizon'].unique():
        # Extract the numeric part from the horizon string to construct the 'lag_retX' column name
        horizon_number = int(''.join(filter(str.isdigit, horizon)))
        lag_ret_column = f'lag_ret{horizon_number}'

        if lag_ret_column in evaluation_i.columns:
            # Calculate the mean of the 'lag_retX' column for the current horizon
            lag_ret_mean = evaluation_i[evaluation_i['horizon'] == horizon][lag_ret_column].mean()
            # Add the mean of 'lag_retX' to the group means DataFrame under a new column 'Actual'
            group_means.loc[horizon, 'Actual'] = lag_ret_mean
        else:
            print(f"Column {lag_ret_column} does not exist for horizon {horizon}")

    print(group_means)
```
